 # Décomposition en rampes : Ch2-Ex2.1 b)
 

Le signal b) (**sans le Dirac**) de l'exercice [Ch2-Ex2.1-b](https://moodle.insa-toulouse.fr/mod/resource/view.php?id=24741#page=35)

Au début on peut penser à :
  * t<0 pente nulle
  * 0<t<1 : pente de 1 donc $r(t)$
  * 1<t<3 : pente de $-2.r(t-1)$
  * 3<t : pente nulle donc $0.r(t-3)$
Ce qui donnerait 

```octave
plot(t,r(t)-2*r(t-1)+0*r(t-2),'o-');
grid on;
```

Les signaux s'ajoutent ! donc l'orsque l'on met une rampe $r(t)$ à l'instant 0 et si l'on veut "enlever cette rampe" à partir de $t=1$, il faut tout bonnement *casser la pente* en ajoutant une rampe inverse de même amplitude $-r(t-1)$.

Il faut donc raisonner plutôt ainsi :
  * t<0 : Laisser la pente à 0
  * 0<t<1 : Augmenter la pente 1 soit $+r(t)$
  * 1<t<3 : La pente vaut 1 et on veut -2 donc il faut diminuer la pente de 3 soit $-3.r(t-1)$
  * etc.
Ce qui donne

```octave
plot(t,r(t)-3*r(t-1),'o-');
grid on;
hold on; % permet d'ajouter une courbe par dessus
%% A VOUS DE MODIFIER et d'exécuter (MAJ+ENTER)
plot(t,r(t)-3*r(t-1)+r(t-3),'r+-'); % ce signal est en rouge 'r' 
```

# Dérivées de discontinuités
Reprenons la rampe avec la discontinuité du [Ch3-Ex1-e](https://moodle.insa-toulouse.fr/mod/resource/view.php?id=24741#page=53)

## Dérivons à l'ancienne

Si on dérive cette fonction on détecte :
 * t<1 pente nulle
 * 1<t<2 pente de 2
 * 2<t pente nulle
 
       
On pense donc que la dérivée $e'$ du signal $e$ est sous forme d'échelon $e' = 2.u(t-1)-2.u(t-2)$.

Utilisont l'opérateur d'intégration `OpI`pour voir si en intégrant la dérivée on retrouve bien la fonction.  
    Pour cela on défini la fonction eprime pour la donner à l'opérateur

```octave
eprime = @(t) 2*u(t-1)-2*u(t-2) ; % fonction de la variable t en matlab
plot(t,eprime(t),'r+-'); hold on;
plot(t,OpI(eprime)(t),'ko-')
```

Et oui ! Les outils classique de 1A disent bien que toute intégrale est une fonction continue !

> Pas moyen de faire des intégrales discontinues sans ajouter des Diracs qui sont une généralisation du concept de fonction.

Regardons à quoi ressemble la dérivée de $e$ calculée numériquement avec un échelon rendu continu mais pentu !

Pour cela on défini la fonction e (réponse à l'exercice) pour pour pouvoir y appliquer l'opérateur dérivée `OpD`

```octave
e=@(t) 2*r(t-1) -2*r(t-2) - 2 * u(t-2)
subplot(2,1,1) % Deux lignes de graphiques, 1 colonne, on se me tdans le premier (celui du haut)
plot(t,e(t))
subplot(2,1,2) % idem mais graphique 2 (celui du bas)
plot(t,eprime(t),'k.',t,OpD(e)(t))
legend(["derivee calculee classique";"derivee calculee numeriquement"])
```

On voit que comme $e$ est discontinue (saut de -2 à t=2 due à l'échelon $-2u(t-2)$)  et bien dans la dérivée il y a un saut négatif à l'infini.

> il manque la dérivée de $-2.u(t-2)$ qui est $-2.\delta(t-2)=-2\delta_2$

Vérifions numériquement en y ajoutant le Dirac en 2 (fonction `delta_0`) et en intégrant avec `OpI`

```octave
eprimedirac = @(t) eprime(t) - 2*delta_0(t-2);
plot(t,OpI(eprimedirac)(t),'o-')
```

La dérivée du signal $e$ est donc dans l'espace des fonction généralisées (ou distribution)

$e' = 2.u(\bullet-1) \;-\; 2.u(\bullet-2) \;-\; 2.\delta_2$
