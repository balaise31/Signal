## Faire le point
---

Maintenant que l'on sait analyser une série donnée et trouver ses propriétés :
- période
- valeur moyenne / composante continue
- parité (avec ou sans composante continue)
- fréquence des composante et somme partielle
- étirement en amplitude de la fonction (coefs proportionnels à l'amplitude)
- étirement dans le temps de la fonction (période dépendante de $T_0$ mais pas les coefs )

Il faut savoir calculer les coefficients d'une série connaissant uniquement la fonction périodique $f$, comme démarré dans l'exercice 2. 

Pour cela gardez la vision d'espace vectoriel (voir [cette partie du récap. sur série](notebooks/intro_series.ipynb#Calcul-des-coefficients-de-Fourier)) et rappelez que 

> Pour trouver le coefficient $c(n)$ de $\vec{g}$ devant une composante $\vec{W_n}$
>
>  $g(t) = a_0 + \ldots + c(n) \underbrace{e^{i\,2\pi n t}}_{w_n(t)} + \ldots $
>
> il suffit de faire une projection du signal sur la composante avec le produit scalaire :
> 
> $c(n) = \left\|\text{proj}_{/W_n}\left(\vec{g}\right)\right\| \overset{\text{base ortho.}}{=} \frac{< e^{i2\pi n \bullet}, g>_p}{\|e^{i2\pi n \bullet}\|^2} \overset{\text{base normée}}{=} < e^{i2\pi n \bullet}, g>_p \overset{\text{g périodique}}{=} \frac{1}{T_0}\int_{0}^{T_0}g(t).\overline{e^{i2\pi n t}}\;dt$

La formule de recomposition est évidente : 

> $\vec{g}=\ldots \underbrace{<\vec{w_{-1}}, \vec{g}>}_{c(-1)}.\vec{w_{-1}}+ \underbrace{<\vec{w_{0}}, \vec{g}>}_{c(0)}.\vec{w_{0}}+ \underbrace{<\vec{w_{1}}, \vec{g}>}_{c(1)}.\vec{w_{1}}+\ldots$
> $\vec{g}=\sum_{n=-\infty}^{+\infty}\underbrace{<\vec{w_{n}}, \vec{g}>}_{c(n))}.\vec{w_{n}}$

**Attention** la base de vecteur $\left( \vec{W_n}\right)_{n\in\mathbb{Z}}$ est infinie certes, mais il suffit de calculer **une seule intégrale de produit scalaire pour un $n\in\mathbb{Z}$ quelconque** pour obtenir la formule des coefficients $c(n)$ qui est en fait **une suite numérique**

Cette vision est expliquée dans le récapitulatif sur les séries [ici pour les c(n)](notebooks/intro_series.ipynb#Coefs-de-Fourier-avec-une-base-imaginaire-norm%C3%A9e-:-c(n)) et [là pour les a(n), b(n)](notebooks/intro_series.ipynb#Coefs-de-Fourier-avec-une-base-r%C3%A9elle-:-a(n)-et-b(n)) (attention au facteur $2$ dans la formule, car base non normée)

# Calculs et simplifications
---


## La période doit disparaître !
---
> $w_0.T_0 = 2\pi$  et $f_0.T_0=1$ et tout disparaît !

Quand on intègre sin, cos ou $e^{i\,n\omega_0 t}$ sur une période $T_0$ alors on se retrouve dans la primitive à remplacer $t$ par des valeurs proportionnelles à $T_0$ (disons $a T_0$) ce qui donne des termes :

$ n\omega_0\,\underbrace{a.T_0}_{t} $ qui devient $n \frac{2\pi}{T_0}.a.T_0 = n.a.2\pi$

On obtient ainsi souvent des formules pour les coefficients qui sont assez trigonométriques du genre :

$ c(n) = \frac{A}{n}e^{i.n.a.2\pi}$ ou $\frac{A}{n}\cos(i.n.a.2\pi)$ avec `a` un nombre souvent rationnel.

>  **On doit ainsi faire disparaître $T_0$, $f_0$ et $\omega_0$ de l'expression des coefficients** 



## Simplification trigonométrique complexes
---


Les valeurs $e^{i.n.a.2\pi}$ ou $\cos(i.n.a.2\pi)$ prennent des valeurs remarquables de manière périodique.

**On doit simplifier ces valeurs trigonométriques périodiques et extraire des sous-suites (paires, impaires, modulo 3 etc.) de coefficients**

Cela dépend des  valeurs de $a$ souvent entières ou fractionnaires qui se simplifient souvent selon la parité de $n$ :

a  | forme $e^{i.n.a.2\pi}$| simplifiée avec $n,p \in \mathbb{Z}$
---|---|---
1 ou entier $m$ | $e^{i. n.m.2\pi}$ | $1 ,\quad \forall n$
$\frac{1}{2}+m$ | $e^{i. n.m.2\pi}.\underbrace{e^{i. n.\frac{2\pi}{2}}}_{\pm 1}$ | $(-1)^n ,\quad \forall n$
$\frac{1}{4}+m$ | $e^{i. n.m.2\pi}.\underbrace{e^{i. n.\frac{2\pi}{4}}}_{\pm 1 \text{ou} \pm i}$ | $(-1)^p$ pour $n=2p$ et $i.(-1)^{p}$ si $n=2p+1$
$\frac{3}{4}+m$ | $e^{i. n.m.2\pi}.\underbrace{e^{i. n.2\pi\frac{-1}{4}}}_{\pm 1 \text{ou} \pm i}$ | $(-1)^p$ pour $n=2p$ et $-i.(-1)^{p}$ si $n=2p+1$
$\frac{p}{q}+m$ | $e^{i. n.m.2\pi}.\underbrace{e^{i. n.2\pi\frac{p}{q}}}_{\text{racines de} x^k=1}$ | racines qièmes de l'unité 

Pour cela **bien dessiner le cercle unité, indiquer le point pour n=0 puis 1 etc. et repérer les valeurs identiques au signe près**

![racines](gifs/anim_racines_unite.gif)

Les coefficients sont alors définis par morceaux, par exemple pour $c(n)= \frac{A}{n} e^{i n \frac{2\pi}{3}}$ et $c(0)=v_m$ :

$c(n) = \left\{ \begin{array}{lll}  \frac{A}{n} & \text{ pour } n = 3p, & p\in\mathbb{Z}^* \\ \frac{A}{n}.\frac{-1+i\sqrt{3}}{2} &\text{ pour } n = 3p+1, & p\in\mathbb{Z} \\ \frac{A}{n}.\frac{-1-i\sqrt{3}}{2} &\text{ pour } n = 3p+2 , & p\in\mathbb{Z} \\ v_m &\text{ pour } n = 0 & \end{array}  \right.$

Si on décompose en $a(n)$ et $b(n)$ nous avons deux cas :

$a(0) = c(0) = v_m$

$a(n) = c(-n) + c(n) = 2\mathcal{R}\left(c(n)\right) = \left\{ \begin{array}{lll}  \frac{2A}{n} & \text{ pour } n = 3p, & p\in\mathbb{N}^* \\ \frac{-A}{n} & \text{ sinon } & \end{array} \right.$

$b(n) = c(-n) - c(n) = -2\mathcal{I}\left(c(n)\right) = \left\{ \begin{array}{lll}  \frac{-A\sqrt{3}}{n} & \text{ pour } n = 3p+1, & p\in\mathbb{N} \\ \frac{A\sqrt{3}}{n} & \text{ pour } n = 3p+2, & p\in\mathbb{N} \\ 0 & \text{ sinon } & \end{array} \right.$

Vérifions avec Octave :


```python
n = 1:6;
cn  = 1./n .* exp(i*n*2*pi/3);
an = 2*real(cn);
bn = -2*imag(cn);

[ an ; 
% formule a(n) pour
% n=1   n=2   n=3(p=1) n=4   n=5  n=6(p=2)
% 3p+1  3p+2   3p      3p+1  3p+2   3p
-1/1, -1/2,  2/3,    -1/4  -1/5 2/6   
] 
disp ("ça colle pour les a(n)!")

[bn ;
% formule b(n) pour
% n=1   n=2   n=3(p=1) n=4   n=5  n=6(p=2)
% 3p+1  3p+2   3p      3p+1  3p+2   3p
[ -1/1,  1/2,   0,    -1/4    1/5   0   ] *sqrt(3) 
]
disp ("ça colle pour les b(n)!")
```

# Sinusoïde cachée
---

Souvent une primitive par exemple $\left[e^{i n\omega_0 t}\right]_{aT_0}^{b^T_0}$ mène à une somme de deux exponentielles
$e^{i 2\pi n b} - e^{i 2\pi n a}$

voire pour $a=0$ à :

> $e^{i 2\pi b}-1$

Ne pas hésiter à factoriser par "l'argument médian" pour se rapporter à une formule d'Euler et obtenir un sinus ou cosinus.

Ce qui donne pour $e^{i 2\pi b n }-1 = e^{i 2\pi b n}-e^{i 2\pi 0 n}$ un argument médian de $i 2\pi \frac{b+0}{2} n = i \pi b n$ :

> $e^{i 2\pi n b} - 1 = e^{i \pi b n} . 2i .  \frac{e^{i \pi b n}-e^{-i \pi b n}}{2i} = 2i.e^{i 2\pi b n}. \sin(\pi b n)$


Dans le cas plus général, pour $e^{i 2\pi n b} - e^{i 2\pi n a}$, l'argument médian est $i 2\pi\frac{a+b}{2}n$ et on obtient

$e^{i 2\pi n b} - e^{i 2\pi n a} = e^{i 2\pi\frac{a+b}{2}n} . 2i .  \frac{e^{i 2\pi\frac{b-a}{2}n}-e^{-i 2\pi\frac{b-a}{2}n}}{2i} = 2i.e^{i 2\pi\frac{a+b}{2}n}. \sin(2\pi\frac{b-a}{2}n)$

# Faire le lien avec Laplace (exo 2)
---


L'esprit est similaire et les formules aussi **mais** les espaces sont très différents. Toutefois, on peut transporter des transformées connues et les principaux théorèmes de Laplace vers Fourier et inversement car :


$\frac{1}{T_0}\mathrm{L}\left[f.\Pi_{[0 ; T_0]}\right](p=i n \omega_0) \quad = \quad \text{SdF}\left[f\right](n)$

> on fait le Laplace du signal $f$ sur une période (nul ailleurs), on divise par $T_0$ et on remplace $p$ par $i n \omega_0$

On récupère ainsi **toutes les formules de transformées de Laplace connues en séries de Fourier connues** et évite des calculs.

### Dans l'exo 2


On a vu [exo 1.g](#ch3_s4_laplace.ipynb#%C3%80-faire) la transformée de Laplace d'un lobe positif de sinus isolé (somme de deux sinus causaux, mais l'un avec un retard d'une demi-période). C'est le signal `f` de l'exercice 2 à des coefficients près. On peut donc retrouver $F(p)$ avec les propriétés de linéarité et du retard :

$F(p) = L[sin(\alpha\bullet)u(\bullet)](p) + L[sin(\alpha(\bullet-\tau)).u(\bullet-\tau)](p) = \underbrace{(1+e^{-\tau p})}_{\text{linéarité + th.. retard}} \underbrace{\frac{1}{|\alpha|}L[sin(\bullet).u(\bullet)](\frac{p}{\alpha})}_{\text{Th. changement échelle}} $

On retrouve la transformée du sinus en se rappelant qu'il s'agit de deux pôles complexes conjugués, que le tout doit être homogène à $1/p$, et que la dérivée de sin donne cos.

Le théorème de changement d'échelle s'obtient par un changement de variable dans la formule de Laplace [voir le poly 3.3.3 page 44](https://moodle.insa-toulouse.fr/mod/resource/view.php?id=24741#page=44) **attention formule valable pour k>0, refaire la démo pour k<0**

On a alors en prenant $\tau=T_0$ **sans se tromper sur la période $T_0$** de $|sin(\alpha\bullet)|$ qui est de fois plus courte que celle de $sin(\alpha\bullet)$ ! :

$c(n) = \frac{1}{T_0}F(p=i n \omega_0)$

On doit trouver le même résultat, mais il faut savoir simplifier les expressions !

# Exercice 1.1 et 1.2
---

On peut donc associer une fonction périodique $f$ (de variable t temporelle) à une suite de nombre $c(n)$ (où la variable n est liée à la fréquence).

On dit que ces coefficients de Fourier sont une vision fréquentielle du signal ou le spectre : on peut alors tracer la série réelle associée comme un signal dépendant de la fréquence.

L'animation suivante permet de voir le passage de la fonction $f$ (en rouge) du domaine temporel à la suite de coefficients $\hat{f}$ (en bleu) du domaine fréquentiel (les c(n)) :

<img src=gifs/anim_decompo_a_frequentiel.gif width=400>

Dans cet exercice on vous donne les spectres $\hat{f}$ des signaux et on a la relation 

$\hat{f}(f=n.f_0) = c(n)$ et donc par exemple $S_1(f=nf_0) = C_{f_1}(n)$ pour le signal $S_1$ de l'exercice.

## 1.1

Pour relever les aberrations :
* les unités se retrouvent par homogénéité de la série de Fourier ($f$ est en Volts, le temps en secondes donc ...)
* le filtre est linéaire donc la sortie à une fréquence pure est la composante de cette fréquence pure en entrée multipliée par le gain (phaseur ou fonction de transfert à cette fréquence) du système à cette fréquente. Ce gain peut être nul, mais pas infini ! **un filtre peut supprimer une composante à une fréquence, mais ne peut pas en faire apparaître une par magie !** Le système n'est alors pas linéaire, car il ajoute quelque chose (il est affine). On dit que ce n'est pas un système passif, car il est source d'un signal en plus et nécessite donc une source d'énergie qui n'est pas l'entrée (système actif). 

## 1.2

On vous donne donc les modules et phases des $c(n)$ pour différentes fréquences ! Pour trouver la relation entre $f$ et $n$ il faut trouver la période.
* Trouvez la période du signal $S_1$ et remarquez que ce n'est pas la même que celle de $S_2$
* On peut dire que $S_2$ est de période deux fois plus lente ! Cela revient à mettre un coef sur deux de $S_2$ à zéro pour avoir la même période que $S_1$. Ou simplement dire, par exemple qu'à 100Hz $S_1$ à une composante, mais $S_2$ n'en a pas et donc mettre son coef à 0.
* La série de la somme est la somme des séries, n'oubliez qu'il y a une phase !
* Pour la puissance moyenne, c'est simple la base complexe est orthonormée donc il suffit d'ajouter les coefs au carré comme pour les vecteurs dans une base ! 

La base $\left(w_n = e^{i 2\pi f_0 \bullet}\right)_{n\in\mathbb{Z}}$ est orthonormée, donc un signal $g$ se décompose : 

> $\vec{g}=\ldots \underbrace{<\vec{w_{-1}}, \vec{g}>}_{c(-1)}.\vec{w_{-1}}+ \underbrace{<\vec{w_{0}}, \vec{g}>}_{c(0)}.\vec{w_{0}}+ \underbrace{<\vec{w_{1}}, \vec{g}>}_{c(1)}.\vec{w_{1}}+\ldots$

> $\vec{g}=\sum_{n=-\infty}^{+\infty}\underbrace{<\vec{w_{n}}, \vec{g}>}_{c(n)}.\vec{w_{n}}$

Sa **puissance moyenne sur une période (Watts)** est la norme issue du produit scalaire :
> $\|\vec{g}\|^2=<\vec{g}, \vec{g}>=\frac{1}{T_0}\underbrace{\int_0^{T_0} \overset{\text{puissance instantannée (Watts)}}{g(t).\overline{g(t)}}dt}_{\text{énergie d'une période (Joules)}}$

Tout se passe pour cet espace infini comme pour une base Euclidienne finie avec la norme :
> $\|\vec{g}\|^2 \overset{\text{ortho.}}{=} \sum_{n=-\infty}^{+\infty}\|\underbrace{<\vec{w_{n}}, \vec{g}>}_{c(n)}.\vec{w_{n}}\|^2\overset{\text{normée}}{=} \sum_{n=-\infty}^{+\infty}\|c(n)\|^2$



**Attention les c(-n) ne sont pas représentés et leur puissance moyenne n'est pas nulle : ne pas oublier les coefficients négatifs !**
